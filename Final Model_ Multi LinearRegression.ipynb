```python
# Import dependencies
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from matplotlib import style
import os
from sklearn.linear_model import LinearRegression
```

```python
# Read in the data
url_climate_data_complete = 'https://raw.githubusercontent.com/Ricardolpz99/Turn-Green/main/Database/Clean/Climate_data_all_variables.csv'
df = pd.read_csv(url_climate_data_complete)
# Rename Colums Annual CO2 emissions (per capita) and Renewables (% sub energy)
df= df.rename(columns={"Annual CO2 emissions (per capita)":"Emissions","Renewables (% sub energy)":"Renewables","Renewable electricity per capita (kWh)": "Renewable per capita","Solar Generation - TWh": "Solar Generation","Wind Generation - TWh": "Wind Generation","Hydro Generation - TWh": "Hydro Generation","Geo Biomass Other - TWh": "Geo Biomass Other"})

```

```python
enty = df['Entity'].to_list()
enty
```

```python
# df = df.loc[df ['Entity'] == '']
```

```python
# Def input variable
X = df[['Renewables','Solar Generation', 'Wind Generation','Hydro Generation', 'Geo Biomass Other']]
```

```python
id_df =df[['Entity', 'Code','Year','Emissions']]
```

```python
# The shape of x , with a single feature
X.shape
```

```python
# Assing Target variable
y = df.Emissions
```

```python
# Create an instance of linear regression model
model = LinearRegression()
```

```python
# fitting step
model.fit(X, y)
```

```python
#Generate predictions
y_pred = model.predict(X)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
r_sq = model.score(X, y)
print('coefficient of determination:', r_sq)

# Print the Intercept:
print('intercept:', model.intercept_)

# Print the Slope:
print('slope:', model.coef_) 

# # Predict a Response and print it:
# y_pred = model.predict(X)
# print('Predicted response:', y_pred, sep='\n')
```

```python
# Place the preductions in a Df
predictions_df = pd.DataFrame(y_pred, columns = ['Predictions'])
```

```python
# Resetting a dataframe index with .reset_index()
id_df = id_df.reset_index()
predictions_df = predictions_df.reset_index()
X = X.reset_index()
```

```python
# Drop a Pandas Dataframe index with `.reset_index()
id_df = id_df.reset_index(drop=True)
predictions_df = predictions_df.reset_index(drop=True)
X = X.reset_index(drop=True)
```

```python
#Drop Index
id_df= id_df.drop(columns= ['index'])
predictions_df= predictions_df.drop(columns= ['index'])
X= X.drop(columns= ['index'])
```

```python
#Join Entity, code, year, emissions columns to predictions
predictions_df = id_df.join(predictions_df)
```

```python
#Join predictions to input data
final_df = predictions_df.join(X)
final_df
```

```python
# Convert Df to CSV
final_df.to_csv('Resources/final.csv')
```
