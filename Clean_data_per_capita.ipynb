```python
# Add the Pandas dependency.
import pandas as pd
import numpy as np
```

```python
url_co2 = 'https://raw.githubusercontent.com/Ricardolpz99/Turn-Green/main/Database/co-emissions-per-capita.csv'
CO2_emissions_df = pd.read_csv(url_co2)
```

```python
url_RE_per_capita = 'https://raw.githubusercontent.com/Ricardolpz99/Turn-Green/main/Database/renewable-electricity-per-capita.csv'
RE_per_capita_df=pd.read_csv(url_RE_per_capita)
```

```python
CO2_emissions_df.head(10)
```

```python
RE_per_capita_df.head(10)
```

```python
Climate_data_complete_per_capita_df = pd.merge(CO2_emissions_df, RE_per_capita_df)
Climate_data_complete_per_capita_df.head(60)
```

```python
 len(Climate_data_complete_per_capita_df["Entity"].unique())
```

```python
 len(Climate_data_complete_per_capita_df["Code"].unique())
```

```python
Climate_data_complete_per_capita_df.drop(Climate_data_complete_per_capita_df.index[Climate_data_complete_per_capita_df['Renewable electricity per capita (kWh)'] == 0], inplace=True)
```

```python
Climate_data_complete_per_capita_df.to_csv('./Database/Clean/Climate_data_complete_per_capita.csv')
```
