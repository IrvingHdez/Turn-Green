```python
# Add the Pandas dependency.
import pandas as pd
import numpy as np
from path import Path
```

```python
url_co2 = 'https://raw.githubusercontent.com/Ricardolpz99/Turn-Green/main/Database/co-emissions-per-capita.csv'
CO2_emissions_df = pd.read_csv(url_co2)
```

```python
url_RE = 'https://raw.githubusercontent.com/Ricardolpz99/Turn-Green/main/Database/renewable-share-energy.csv'
RE_df=pd.read_csv(url_RE)
```

```python
# Read the data file and store it in a Pandas DataFrame.
RE_df.head(10)
```

```python
CO2_emissions_df.head(10)
```

```python
# Combine the data into a single dataset.
Climate_data_complete_df = pd.merge(CO2_emissions_df, RE_df)
Climate_data_complete_df.head()
```
